```python
# Задание 1 
def sum_distance(from_val, to_val):
    if from_val > to_val:
        from_val, to_val = to_val, from_val
    return sum(range(from_val, to_val + 1))
    
print(sum_distance(1, 5))   # 15 (1+2+3+4+5)
print(sum_distance(5, 1))   # 15 (1+2+3+4+5)
print(sum_distance(3, 3))   # 3
```

```python
# Задание 2 
def trim_and_repeat(string, offset=0, repetitions=1):
    if offset >= len(string):
        return ""
    if repetitions < 1:
        return ""
    trimmed = string[offset:]
    return trimmed * repetitions

print(trim_and_repeat("hello", 2, 3))      # "llo" * 3 = "llollollo"
print(trim_and_repeat("hello", 10, 2))     # "" (offset больше длины строки)
print(trim_and_repeat("hello", 1, 0))      # "" (repetitions меньше 1)
print(trim_and_repeat("hello", 0, 2))      # "hellohello"
```
